####Chi-Square Test of Independence (Check independency for Categorical Data)

####Check 'Overweight_Obese_Family' and 'Class' are likely to be related or not

```python
# H0: The two categorical variable are independent
# H1: The two categorical variable are dependent

table = pd.crosstab(df['Overweight_Obese_Family'].replace({1:'Yes',0:"No"},inplace = False),df['Class'].replace({1:"Underweight",2:'Normal',3:'Overweight',4:'Obesity'},inplace =False))
print(table)

```

```python
#Observed Value

observed_value = table.values
observed_value
```

```python
# Get Chi-squared value, p-value, Degree of Freedom and Expected

chi2, p, dof, expected = stats.chi2_contingency(observed_value)

print("Chi-squared:", chi2)
print("p-value:", p)
print("Degrees of freedom:", dof)
print("Expected frequencies:\n", expected)
```

```python

alpha = 0.05

critical_value = stats.chi2.ppf(q = 1-alpha, df= dof)

print(critical_value)
```

Since 2 = 180.39 > 7.81, there is sufficient evidence to indicate that a dependence between "Class" and
"Overweight_Obese_Family" at 5% significant level

```python
from sklearn.feature_selection import chi2
import pandas as pd
import scipy.stats as stats

# Exclude irrelevant columns and target column
a = df.drop(columns=['Age', 'Height', 'Class','Age_Group','Height_Group'], axis=1)
b = df['Class']

# Perform Chi-Square test
chi_scores = chi2(a, b)
chi_statistics = chi_scores[0]
p_values = chi_scores[1]

# Degrees of freedom (df = unique categories in each column - 1)
degrees_of_freedom = [len(df[col].unique()) - 1 for col in a.columns]

# Critical values at 5% confidence level
critical_values = [stats.chi2.ppf(0.95, df) for df in degrees_of_freedom]

# Hypothesis results
hypotheses = ['Dependent' if p < 0.05 else 'Independent' for p in p_values]

# Create a DataFrame to display the results
results_df = pd.DataFrame({
    'Feature': a.columns,
    'Chi-Square Statistic': chi_statistics,
    'P-Value': p_values,
    'Degrees of Freedom': degrees_of_freedom,
    'Critical Value (5%)': critical_values,
    'Hypothesis': hypotheses
})
```

```python
results_df.sort_values(by='P-Value', ascending=True)
```

From analysis above, there is sufficient evidence to indicate that features 'Type_of_Transportation_Used'
 'Consumption_of_Fast_Food', 'Physical_Excercise', 'Overweight_Obese_Family', 'Frequency_of_Consuming_Vegetables', 'Smoking', 'Calculation_of_Calorie_Intake', 'Number_of_Main_Meals_Daily', 'Sex', 'Liquid_Intake_Daily,and 'Food_Intake_Between_Meals' are dependent with 'Class' at 1% significant level

 However, at 5% significant level, there is sufficient evidence to indicate that 'Type_of_Transportation_Used'
 'Consumption_of_Fast_Food', 'Physical_Excercise', 'Overweight_Obese_Family', 'Frequency_of_Consuming_Vegetables', 'Smoking', 'Calculation_of_Calorie_Intake', 'Number_of_Main_Meals_Daily', 'Sex', 'Liquid_Intake_Daily, 'Food_Intake_Between_Meals', and 'Schedule_Dedicated_to_Technology' are dependent with 'Class'




###Analysis of Variance (ANOVA)

####to test whether the means of a continuous variable are significantly different across the categories of a categorical variable

```python
#Null Hypothesis :The means of the continuous variable are the same across all categories.
#Alternative Hypothesis : At least one category has a different mean.

print(df['Class'].unique())

import scipy.stats as stats
# Perform ANOVA
anova_result = stats.f_oneway(
    df[df['Class'] == 1]['Age'],
    df[df['Class'] == 2]['Age'],
    df[df['Class'] == 3]['Age'],
    df[df['Class'] == 4]['Age'])

print(f"ANOVA F-statistic: {anova_result.statistic:.2f}")
print(f"p-value: {anova_result.pvalue:.4f}")
```

From result above, there is sufficient evidence to indicate that Age differ significantly between different weight class at 0.0001% significant level

```python
def model_training(X, y, model, sampler = None):
  skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

  fold_accuracies = []

  for fold, (train_idx, test_idx) in enumerate(skf.split(X, y), 1):
      X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
      y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

      if sampler is not None:
          if isinstance(sampler, dict):
            for name, method in sampler.items():
                print(f"Applying {name}...")
                X_train, y_train = method.fit_resample(X_train, y_train)
          else:
            X_train, y_train = sampler.fit_resample(X_train, y_train)

      # Train the model
      model.fit(X_train, y_train)

      # Predict and evaluate
      y_pred = model.predict(X_test)
      accuracy = accuracy_score(y_test, y_pred)
      fold_accuracies.append(accuracy)

      print(f"Fold {fold}: Accuracy = {accuracy:.2f}")

  # Average accuracy across folds
  print(f"Average Accuracy: {np.mean(fold_accuracies):.2f}")

  print(classification_report(y_test, y_pred))
  ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
```

```python
from sklearn.model_selection import RandomizedSearchCV
from imblearn.pipeline import Pipeline

# Define the RandomForestClassifier model
model = RandomForestClassifier(random_state=42, class_weight='balanced')
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
oversampler = RandomOverSampler(sampling_strategy = {1: 300})

pipeline = Pipeline([
    ('oversampler', oversampler),
    ('model', model)
])

# Define the param_grid (now param_distributions) with valid options for max_features
param_grid = {
    'model__n_estimators': [100, 200, 300],  # Number of trees in the forest
    'model__max_depth': [10, 20, 30, None],  # Maximum depth of each tree
    'model__min_samples_split': [2, 5, 10],  # Minimum number of samples required to split an internal node
    'model__min_samples_leaf': [1, 2, 4],    # Minimum number of samples required to be at a leaf node
    'model__max_features': ['sqrt', 'log2', None]  # Valid values for max_features
}

# Randomized search with 5-fold cross-validation
random_search = RandomizedSearchCV(
    estimator=pipeline,
    param_distributions=param_grid,  # The parameter grid to search
    n_iter=50,  # Number of random combinations to test
    scoring='f1_weighted',  # Use F1 weighted score for evaluation
    cv=skf,  # 5-fold cross-validation
    verbose=1,  # Print detailed messages
    random_state=42  # Ensure reproducibility
)

# Fit the model with the training data
random_search.fit(X_onehot, y_onehot)

# Print the best parameters found during the search
print("Best Parameters:", random_search.best_params_)
```
